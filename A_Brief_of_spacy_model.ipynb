```python
import sys
from pathlib import Path

import pandas as pd

import spacy
from spacy import displacy
from textacy.extract import ngrams, entities
```

```python
#Download the spacy model
%%bash
python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load('en_core_web_sm') 
```

```python
with open("example.txt", "r",encoding='UTF-8') as f:
    lines = f.readlines()
    aa =' '.join(lines)

    print(aa)
```

```python
doc = nlp(aa)
sentences = [s for s in doc.sents]
len(sentences)
sentences[0]
```

```python
temp = pd.DataFrame([[t.text, t.lemma_, t.is_alpha, t.is_stop] for t in doc], columns = ['text', 'lemma', 'is_alpha', 'is_stop'])
temp.head(30)
```

```python
#identify the entity words
for t in sentences[0]:
    if t.ent_type_:
        print('{} | {} | {}'.format(t.text, t.ent_type_, spacy.explain(t.ent_type_)))
```

```python
#Value count
entities = [e.text for e in entities(doc)]
pd.Series(entities).value_counts().head()
```
